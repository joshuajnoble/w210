```python
import os

# # assign path
path, dirs, files = next(os.walk("/Users/joshnoble/Desktop/pulses"))
file_count = len(files)
sorted_files = sorted(files)
sorted_files
# # create empty list
dataframes_list = []
 
# # append datasets to the list
for i in range(file_count):
    temp_df = pd.read_csv("~/Desktop/pulses/"+sorted_files[i])
    dataframes_list.append(temp_df)
```

```python
pulse_df = pd.DataFrame()
```

```python
def process_df( in_df, survey, week, geo ):
    tmp = {}
    tmp['survey'] = survey
    tmp['week'] = week
    tmp['geography'] = geo
    tmp['white'] = len(in_df[(in_df['RRACE'] == 1)])
    tmp['black'] = len(in_df[in_df['RRACE'] == 2])
    tmp['asian'] = len(in_df[in_df['RRACE'] == 3])
    tmp['no_hs'] = len(in_df[in_df['EEDUC'] == 1])
    tmp['some_hs'] = len(in_df[in_df['EEDUC'] == 2])
    tmp['hs'] = len(in_df[in_df['EEDUC'] == 3])
    tmp['some_college'] = len(in_df[in_df['EEDUC'] == 4])
    tmp['associates'] = len(in_df[in_df['EEDUC'] == 5])
    tmp['bachelors'] = len(in_df[in_df['EEDUC'] == 6])
    tmp['graduate'] = len(in_df[in_df['EEDUC'] == 7])
    tmp['married'] = len(in_df[in_df['MS'] == 3])
    tmp['un_married'] = len(in_df[in_df['MS'] != 3])
    tmp['kids_in_hh'] = len(in_df[(in_df['THHLD_NUMKID'] > 0)])
    if( 'WRKLOSSRV' in in_df.columns):
        tmp['job_loss_yes'] = len(in_df[(in_df['WRKLOSSRV'] == 1)])
        tmp['job_loss_no'] = len(in_df[(in_df['WRKLOSSRV'] == 2)])
    else:
        tmp['job_loss_yes'] = len(in_df[(in_df['WRKLOSS'] == 1)])
        tmp['job_loss_no'] = len(in_df[(in_df['WRKLOSS'] == 2)])
        
    if( 'EXPNS_DIF' in in_df.columns):
        tmp['expenses_no_difficulty'] = len(in_df[(in_df['EXPNS_DIF'] == 1)])
        tmp['expenses_a_little_difficulty'] = len(in_df[(in_df['EXPNS_DIF'] == 2)])
        tmp['expenses_some_difficulty'] = len(in_df[(in_df['EXPNS_DIF'] == 3)])
        tmp['expenses_very_difficult'] = len(in_df[(in_df['EXPNS_DIF'] == 4)])
    else:
        tmp['expenses_no_difficulty'] = -1
        tmp['expenses_a_little_difficulty'] = -1
        tmp['expenses_some_difficulty'] = -1
        tmp['expenses_very_difficult'] = -1
        
    if( 'RENTCUR' in in_df.columns):
        tmp['rent_cur_yes'] = len(in_df[(in_df['RENTCUR'] == 1)])
        tmp['rent_cur_no'] = len(in_df[(in_df['RENTCUR'] == 2)])
    else:
        tmp['rent_cur_yes'] = -1
        tmp['rent_cur_no'] = -1
        
    if( 'TMNTHSBHND' in in_df.columns):
        tmp['months_behind_avg'] = in_df[(in_df['TMNTHSBHND'] > 0)]['TMNTHSBHND'].mean()
    else:
        tmp['months_behind_avg'] = -1

    if( 'RENTASSIST' in in_df.columns):
        tmp['rental_assistance_recvd'] = len(in_df[(in_df['RENTASSIST'] == 1)])
        tmp['rental_assistance_waiting'] =len(in_df[(in_df['RENTASSIST'] == 2)])
        tmp['rental_assistance_denied'] =len(in_df[(in_df['RENTASSIST'] == 3)])
        tmp['rental_assistance_not_needed'] =len(in_df[(in_df['RENTASSIST'] == 4)])
    else:
        tmp['rental_assistance_recvd'] = -1
        tmp['rental_assistance_waiting'] = -1
        tmp['rental_assistance_denied'] = -1
        tmp['rental_assistance_not_needed'] = -1
    
    if( 'EVICT' in in_df.columns):
        tmp['eviction_likely'] =len(in_df[(in_df['EVICT'] == 1)])
        tmp['eviction_somewhat_likely'] =len(in_df[(in_df['EVICT'] == 2)])
        tmp['eviction_not_likely'] =len(in_df[(in_df['EVICT'] == 3)])
        tmp['eviction_very_unlikely'] =len(in_df[(in_df['EVICT'] == 4)])
    else:
        tmp['eviction_likely'] = -1
        tmp['eviction_somewhat_likely'] = -1
        tmp['eviction_not_likely'] = -1
        tmp['eviction_very_unlikely'] = -1
    
    return tmp
```

```python
for i in range(len(dataframes_list)):
    week = ( int(i) * 3) + 18
    df = dataframes_list[i]
    survey = i+1
    dallas = df[df["EST_MSA"] == 19100]
    d = process_df( dallas, survey, week, "Dallas" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    d = process_df( dallas, survey, week+1, "Dallas" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    d = process_df( dallas, survey, week+2, "Dallas" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    houston = df[df["EST_MSA"] == 26420]
    d = process_df( houston, survey, week, "Houston" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    d = process_df( houston, survey, week+1, "Houston" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    d = process_df( houston, survey, week+2, "Houston" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    texas = df[(df['EST_ST'] == 48) & (df['EST_MSA'].isnull())]
    d = process_df( texas, survey, week, "Texas" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    d = process_df( texas, survey, week+1, "Texas" )
    pulse_df = pulse_df.append(d, ignore_index=True)
    d = process_df( texas, survey, week+2, "Texas" )
    pulse_df = pulse_df.append(d, ignore_index=True)
```
